## Cleaning Data

```python
import re

s = "http\\:www.mynet.com #lateflight @airlines"
s = re.sub("http\S+", "", s).strip()
s
```

```python
s = re.sub("#\S+", "", s)
s
```

```python
stop_words = stopwords.words('english')

for i in ["not", "no"]:
    stop_words.remove(i)
```

```python
def cleaning(data):

    import re

    #1. Removing URLS
    data = re.sub('http\S+', '', data).strip()
    data = re.sub('www\S+', '', data).strip()

    #2. Removing Tags
    data = re.sub('#\S+', '', data).strip()

    #3. Removing Mentions
    data = re.sub('@\S+', '', data).strip()

    #4. Removing upper brackets to keep negative auxiliary verbs in text
    data = data.replace("'", "")  # If you use negative_auxiliary_verbs above, comment this line

    #5. Tokenize
    text_tokens = word_tokenize(data.lower())

    #6. Remove Puncs and number
    tokens_without_punc = [w for w in text_tokens if w.isalpha()]

    #7. Removing Stopwords
    tokens_without_sw = [t for t in tokens_without_punc if t not in stop_words]

    #8. lemma
    text_cleaned = [
        WordNetLemmatizer().lemmatize(t) for t in tokens_without_sw
    ]

    #joining
    return " ".join(text_cleaned)
```

```python
cleaning_text = df2["text"].apply(cleaning)
cleaning_text.head()
```

## Features and Label

## Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = df2["text"]
y = df2["airline_sentiment"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.2,  # If you get bad test results make the propotion fewer
                                                    stratify=y,
                                                    random_state=101)
```

## Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(preprocessor=cleaning, min_df=3)
            # ngram_range=(1, 2) makes word group consist of two words, max_features=1500, 
            # min_df 3-4 or 5 is ok, max_df usage is crucial (if you, %95 or higher is good) because of ignoring some valuable token for our model
X_train_count = vectorizer.fit_transform(X_train)
X_test_count = vectorizer.transform(X_test)

# min_df=3    --> corpusumda 3'den daha az document'da (yorumda) geen tokenleri ignore et.

# max_df=1000 --> corpusumda 1000'den daha fazla document'da (yorumda) geen tokenleri ignore et. (0.95, 1.0)
# max_features=1500 --> corpusumda en sk kullanlan ilk 1500 feature (tokeni) dikkate al dierlerini ignore et.
```

```python
X_train_count.toarray()
```

```python
pd.DataFrame(X_train_count.toarray(),
             columns=vectorizer.get_feature_names_out())
```

## Model Comparisons - Vectorization

```python
from sklearn.metrics import confusion_matrix, classification_report, f1_score, recall_score
```

```python
def eval(model, X_train, X_test):
    y_pred = model.predict(X_test)
    y_pred_train = model.predict(X_train)
    print(confusion_matrix(y_test, y_pred))
    print("- --- -" * 3)
    print("Test_Set")
    print(classification_report(y_test, y_pred))
    print("\n", "- --- -" * 8, "\n")
    print("Train_Set")
    print(classification_report(y_train, y_pred_train))
```

```python
from sklearn.naive_bayes import MultinomialNB, BernoulliNB  # BernoulliNB for binary model
```

```python
nb = MultinomialNB(alpha=10)
nb.fit(X_train_count, y_train)
```

```python
print("NB MODEL\n")
eval(nb, X_train_count, X_test_count)
```

```python
from sklearn.model_selection import cross_validate
from sklearn.metrics import make_scorer
from sklearn.metrics import f1_score, accuracy_score, precision_score, recall_score

scoring = {
    'accuracy':
    make_scorer(accuracy_score),
    'precision-neg':
    make_scorer(precision_score, average=None, labels=["negative"]),
    'recall-neg':
    make_scorer(recall_score, average=None, labels=["negative"]),
    'f1-neg':
    make_scorer(f1_score, average=None, labels=["negative"])
}

model = MultinomialNB(alpha=10)
scores = cross_validate(model,
                        X_train_count,
                        y_train,
                        scoring=scoring,
                        cv=10,
                        return_train_score=True)

df_scores = pd.DataFrame(scores, index=range(1, 11))
df_scores.mean()[2:]
```

```python
from yellowbrick.classifier import PrecisionRecallCurve

viz = PrecisionRecallCurve(MultinomialNB(alpha=10),
                           classes=nb.classes_,
                           per_class=True,
                           cmap="Set1")
viz.fit(X_train_count, y_train)
viz.score(X_test_count, y_test)
viz.show();
```

```python
y_pred = nb.predict(X_test_count)
nb_count_rec_neg = recall_score(y_test,
                                y_pred,
                                labels=["negative"],
                                average=None)
nb_count_f1_neg = f1_score(y_test, y_pred, labels=["negative"], average=None)
nb_AP_neg = viz.score_["negative"]
```
